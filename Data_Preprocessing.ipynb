# Data Preprocessing - Oversampling : SMOTE

```python
#Smote
from imblearn.over_sampling import SMOTE
```

```python
feature = df.iloc[:, 0:30]
target = df.iloc[:,30]
df.columns.values
```

```python
smote = SMOTE()

x_smote, y_smote = smote.fit_resample(feature, target)
x_smote
```
